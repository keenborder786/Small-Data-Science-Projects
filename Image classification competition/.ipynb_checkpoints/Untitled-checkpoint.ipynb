```python
os.chdir(r'C:\Users\MMOHTASHIM\Anaconda3\libs\Small Data Science projects\Small-Data-Science-Projects\Image classification competition')
X=np.load("X_ARR.npy")
y=np.load("y_ARR.npy")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
X_valid,y_valid=X_train[-33:],y_train[-33:]
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',padding='valid',input_shape=(24, 24, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), padding='valid',activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), padding='valid',activation='relu'))
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint("my_keras_model.h5",
save_best_only=True)
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=30,
validation_data=(X_valid, y_valid),
callbacks=[checkpoint_cb])

y_valid
```

```python
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
plt.show()
```
