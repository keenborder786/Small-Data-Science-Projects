<h1>Imports</h1>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import MaxAbsScaler
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import f_regression
from sklearn.decomposition import PCA
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectKBest
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
df.set_index("ID",inplace=True)
df.fillna(value=0,inplace=True)
```

<h3>Randomforest regressor</h3>
I used Randomforest regressor to predict customer future transcation values, as data was sparsed I used PCA to reduce its dimensionality. 

```python
X=np.array(df.drop("target",1))
pca=PCA(n_components=1000)
X_reduced=pca.fit_transform(X)
y=np.array(df["target"])
clf=RandomForestRegressor(n_estimators=100,verbose=10,n_jobs=-1)
clf.fit(X_reduced,y)
```

```python
test=pd.read_csv("test.csv")
```

```python
test.head()
test.reset_index(inplace=True)
```

```python
submission=pd.DataFrame()
submission["ID"]=test["ID"]
submission.set_index("ID",inplace=True)
```

```python
test.set_index("ID",inplace=True)
```

```python
X_test=np.array(test)
X_test=pca.transform(X_test)
submission["target"]=clf.predict(X_test)
```

<h4>Lasso Regression</h4>
Further I also used Lasso regression after selecting best features through mutual info regression and then also normalising the data.

```python
X_test=np.array(test)
X=np.array(df.drop("target",1))
y=np.array(df["target"])
feat = SelectKBest(mutual_info_regression,k=200)
X_tr = feat.fit_transform(X,y)
X_te = feat.transform(X_test)
scaler = MaxAbsScaler()
tr_data = scaler.fit_transform(X_tr)
te_data = scaler.fit_transform(X_te)
reg = Lasso(alpha=0.0000001, max_iter = 10000)
reg.fit(tr_data,y)
```

<h5>Submission to Kaggle</h5>

```python
submission["target"]=reg.predict(te_data)
submission.shape
```

```python
submission.to_csv("submission-2.csv")
```
