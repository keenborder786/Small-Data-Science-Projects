```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib notebook
directory=r"C:\Users\MMOHTASHIM\Downloads"
df=pd.read_csv(directory+"\dataset_Facebook.csv",sep=';',engine='python')
columns=df.columns.tolist()
columns
```

```python
list_media=df.Type.unique()
divided_data={}
data=np.array(df.values.tolist())
for media in list_media:
    temp_data=[]
    for i in data:
        if i[1]==media:
            i=np.delete(i,1)
            temp_data.append(i)
    divided_data[media]=temp_data
divided_data
    

    
```

```python
Photos_data=np.array(divided_data['Photo']).astype(float)
Status_data=np.array(divided_data['Status']).astype(float)
Link_data=np.array(divided_data['Link']).astype(float)
Video_data=np.array(divided_data['Video']).astype(float)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Photos_data[:,13],Photos_data[:,6],Photos_data[:,-1])
ax.set_xlabel("Lifetime People who have liked your Page and engaged with your post")
ax.set_ylabel('Lifetime Post Total Reach')
ax.set_zlabel("Total Interaction")
ax.set_title("Photo Performance")
ax.set_xlim(0,4000)
ax.set_ylim(0,200000)
ax.set_zlim(0,7000)
plt.show()
```
